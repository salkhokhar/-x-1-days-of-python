# Intro to Strings  
String are contained within single('') or double ("") quotes.  


```python
str1 = 'Hello world!'
str2 = "Hello Sal"
print(str1)
print(str2)
```

If there you have to insert a quotation mark within a string there are two methods of doing so.  
1. If you have to use singe quote(') within the string, use double quotes ("") for declaring the string and vice versa
2. You can use backslash(\) just before the quote.

```python
# you can't do this
str1 = 'It's fine' 
str2 = "Press "Enter""
print(str1)
print(str2)
```

```python
# Method 1
str1 = "It's fine" 
str2 = 'Press "Enter"'
print(str1)
print(str2)
```

```python
# Method 2
str1 = 'It\'s fine' 
str2 = "Press \"Enter\""
print(str1)
print(str2)
```

# Combining strings

```python
str1 = "Hello sal."
str2 = "I'm vegeta!!"
# we can combine strings with a simple (+) sign
new_str = str1 + str2
print(new_str) 
```

```python
# we combined it alright. But it doesn't look right without any space between the two sentences
# so let's add the space
new_str = str1 + " " + str2
print(new_str)
```

# Escape characters

```python
# we can also use some escape codes (\', \\, \b, \t, \n, \r, \ooo, \xhh)
# \b is backspace, it remove the character which preceeds it

# \' is used to insert a quote into a string
txt = 'It\'s fine'
print(txt)
# \\ is used to insert a backslash into a string
# it is important to remember that using a single backslash may trigger some escape character
# for eg. address = "E:\New folder" will provide you with the following output
# "E:
# ew folder"
#so if you have to insert backslash in a string use \\
address = "E:\\New folder"
print(address)
# in this case it was the fullstop (.)
new_str = str1 + "\b" + str2
print(new_str)
# \t insert a tab
new_str = str1 + "\t" + str2
print(new_str)
# \n inserts a line break, i.e., it will start a new line
new_str = str1 + "\n" + str2
print(new_str)
# \r passes the cursor to starting point and will overwrite the previously written statement
# to it's effect effect str1 was printed twice as str1 was shorter than str2
new_str = str1 + str1 + "\r" + str2
print(new_str)
# \ooo simply means that if it will convert the octal values passed into ASCII characters
txt = "\110\145\154\154\157"
print(txt) 
# \xhh means that if it will convert the hexadecimal values passed into ASCII characters
txt = "\x48\x65\x6c\x6c\x6f"
print(txt) 
# new_str = str1 + "\f" + str2
# print(new_str)
```

# f-strings
In case we have to insert variables into a string. f-string method just looks very cool and efficient 

```python
name = 'Sal'
age = 23
# going by the boring way
# we also have to first convert the age variable, which is an integer, to string using the str() function
print("Hello, My name is " + name + "and I'm " + str(age) + " years old.")
# Now using the f string formatting mechanism, place the variables in curly brackets {} and before 
# starting the string remember to place the f keyword
print(f"Hello, My name is {name} and I'm {age} years old.")
# if you have to insert a variable in a string at multiple places, we can also use the following format
print("Hello, My name is {0} and I'm {1} years old. Also, I've live for {1} years. Isn't that starnge".format(name, age))
# we can refer to the variables with the index (0,1,...) at which they are stored within the format function
# there is also no need to use the f keywod before the string

# if you decide on using the latter format but have to use each variable only once, then you 
# don't have to mention the index at which the varible is stored, instead just store the
# variable chronologically
print("Hello, My name is {} and I'm {} years old.".format(name, age))

```

# Indexing and Slicing

```python
# Each character in a string can be addressed by its index value
# l is at index 0, e at index 1, t at index 2, " " at index 3 and so on
str1 = "Let it be a somewhat long sentence. Anything will do."
print(str1)
# to address a character by its index we can just use square brackets []
print(f"Addressing a single character: {str1[1]}")
# for by its negative index which starts at -1 and will refer to string from last to first 
print(f"Addressing the last character: {str1[-1]}")
# we can simultaneously address a number of subsequent charcters as follows
# str1[x:y]; x is the staring position and y is the last position upto which the string will be read.
# x is inclusive but y is non-inclusive
print(f"Addressing a bunch of character: {str1[0:5]}")
print(str1[2:10])
print(str1[:10]) # if we have to start from index 0, we can just leaveout x
print(str1[15:]) # if we have to transverse every character till the end, we can leaveout y
print(str1[-7:-5]) # it will provide us wih charcters from index -7 to index -5
print(str1[5:-3]) # it will provide us wih charcters from index 5 to index -3

# we can also introduce steps so that it will skip some charcters
# str1[x:y:z]; z is the step count, it will skip z-1 charcters 
print(f"Skipping some characters: {str1[::2]}")
print(f"Reading the string in reverse: {str1[::-1]}")

```

# Some functions

```python
# to find out the length of a string: len()
print(f"The length of string is: {len(str1)}")

# to convert the string into lowercase and uppercase: lower() & upper()
print(f"The string in lowercase: {str1.lower()}")
print(f"The string in uppercase: {str1.upper()}")

# how many times does some character or substring appears in the string: count()
print(f"The word 'it' is used: {str1.count('it')}")

# to find the location of a character or substring: find()
print(f"The word 'it' appears at: {str1.find('it')}")
# if that character or substring isn't present result will be -1
print(f"The word 'is' appears at: {str1.find('is')}")

# to replace: replace()
print(f"To replace 'long' with 'short': {str1.replace('long', 'short')}")

# we can split the string into a list of strings
print(f"Let's split the string everytime character i appears: {str1.split('i')}")

# we can check the datatype of any varable using : type()
print(f"The type of str1 variable is : {type(str1)}")

```

```python
# there are a lot of attributes and methods that we can use with a function.
# we can see all those with the dir method
print(dir(str1))

# we can get see what these methods an attributes do with the help function: help()
# help() can't take varibles as arguments
# so instead of varible str1 we can pass the argument str which is the datatype of str1
print(help(str))
```

```python
# for getting help on a particular method or attribute 
print(help(str.capitalize))
```
